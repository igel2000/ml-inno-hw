## Policy Start Date

```python
# Разобрать дату старта полису на год, месяц, день и день недели
dataset_df_wo_nan['Policy Start Date'] = pd.to_datetime(dataset_df_wo_nan['Policy Start Date'])

dataset_df_wo_nan['Policy Start Year'] = dataset_df_wo_nan['Policy Start Date'].dt.year
dataset_df_wo_nan['Policy Start Month'] = dataset_df_wo_nan['Policy Start Date'].dt.month
dataset_df_wo_nan['Policy Start Day'] = dataset_df_wo_nan['Policy Start Date'].dt.day
dataset_df_wo_nan['Policy Start Weekday'] = dataset_df_wo_nan['Policy Start Date'].dt.weekday 

dataset_df_wo_nan.drop(columns=['Policy Start Date'], inplace=True)

params["columns_X"].remove("Policy Start Date")
params["cat_columns"].remove("Policy Start Date")
params["columns_X"].append("Policy Start Year")
params["columns_X"].append("Policy Start Month")
params["columns_X"].append("Policy Start Day")
params["columns_X"].append("Policy Start Weekday")

display(dataset_df_wo_nan.sample(7))
```

## Прочие категориальные переменные

```python
print("Набор значений категориальных столбцов:")
for col in params["cat_columns"]:
    #print(f'{dataset_df_wo_nan[col].value_counts()}')
    print(f'{col}: {dataset_df_wo_nan[col].unique().tolist()}')
```

```python
# Столбцы для one-hot-encoding
onehot_columns = ['Gender', "Marital Status", "Occupation", "Location", "Property Type", "Smoking Status"]
# Столбцы для ordinal-encoding
ordered_columns = {"Education Level": ["High School", "Bachelor's", "Master's", "PhD"],
                   "Policy Type": ['Basic', 'Comprehensive', 'Premium'],
                   "Customer Feedback": ['Poor', 'Average', 'Good'],
                   "Exercise Frequency": ['Rarely', 'Weekly', 'Monthly', 'Daily']}
dataset_df_wo_nan, params["columns_X"] = encode_features(src_df=dataset_df_wo_nan, 
                                                         onehot_cols=onehot_columns, onehot_drop='first',
                                                         ordinal_cols=ordered_columns, 
                                                         columns_X=params["columns_X"])
```
