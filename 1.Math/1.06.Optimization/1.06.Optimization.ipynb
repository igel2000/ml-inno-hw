Дана функция $$f({x}_1, {x}_2) = 3{x}^2_1+ 2{x}^2_2+ 4{x}_1{x}_2−5{x}_1+ 6{x}_2$$ 
найти минимум функции с помощью метода градиентного спуска

Частные производные:
$${f}^`_{x_1} = 6{x}_1 + 4{x}_2 - 5$$
$${f}^`_{x_2} = 4{x}_2 + 4{x}_1 + 6$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Определяем функцию и ее производные
def f(x1, x2):
    return 3*x1**2 + 2*x2**2 + 4*x1*x2 - 5*x1 + 6*x2

def df(x1, x2):
    return 6*x1 + 4*x2 -5, 4*x2 + 4*x1 + 6
```

```python
# Градиентный спуск
def gradient_descent_3d(f, df, starting_x, starting_y,learning_rate, num_iterations):
    x, y = starting_x, starting_y
    x_history, y_history, z_history = [x], [y], [f(x, y)]
    for _ in range(num_iterations):
        grad_x, grad_y = df(x, y)
        x -= learning_rate * grad_x
        y -= learning_rate * grad_y
        x_history.append(x)
        y_history.append(y)
        z_history.append(f(x, y))
    return x_history, y_history, z_history
```

```python
# Начальные параметры
starting_x1 = 5
starting_x2 = 5
learning_rate = 0.1
num_iterations = 50
x1_history, x2_history, z_history = gradient_descent_3d(f, df, starting_x1, starting_x2, learning_rate, num_iterations)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x1 = np.linspace(-10, 10, 400)
x2 = np.linspace(-10, 10, 400)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)

ax.plot_surface(X1, X2, Z, cmap='viridis', alpha=0.3)

# Добавление точек градиентного спуска
ax.scatter(x1_history, x2_history, z_history, color='r', s=10, label='Gradient Descent')

ax.set_xlabel('X1 axis')
ax.set_ylabel('X2 axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Gradient Descent Visualization')
ax.legend()

plt.show()

print(f"Минимум достигается при x1={x1_history[-1]:.4f}, x2={x2_history[-1]:.4f}")
print(f"Значение функции в точке минимума = {f(x1_history[-1], x2_history[-1])}:.4f)")
```
